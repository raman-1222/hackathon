```python
links=[link for link in data['image_link']]
import os
import requests
from PIL import Image
from io import BytesIO

def download_images(url_list, folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

    for i, url in enumerate(url_list):
            response = requests.get(url)
            response.raise_for_status() 
            
            image = Image.open(BytesIO(response.content))
            
            
            file_name = f'image_{i+1}.png'
            file_path = os.path.join(folder_path, file_name)
            
          
            image.save(file_path)


save_folder = 'downloaded_images'
download_images(links, save_folder)
```

```python
import os
from PIL import Image
import easyocr

def extract_text_from_images(directory_path):
    reader = easyocr.Reader(['en'])
    text_data= {}
    if not os.path.isdir(directory_path):
        raise ValueError(f"The directory {directory_path} does not exist or is not a directory.")
    
    for idx, image_file in enumerate(os.listdir(directory_path)):
        image_path = os.path.join(directory_path, image_file)
        
        if image_file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image = Image.open(image_path)
            result = reader.readtext(image_path)
            text = " ".join([text for (_, text, _) in result])
            text_data[image_file] = text

    return text_data

directory_path = r'C:\Users\singh\desktop\hackathon\student_resource\downloaded_images'
extracted_texts = extract_text_from_images(directory_path)
print(extracted_texts)
```
